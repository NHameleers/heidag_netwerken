```python
import pandas as pd
```

```python

def get_vaste_staf_namelist(excelfile):
    namen = pd.read_excel(excelfile, sheet_name='Sheet1')

    return list(namen['Volledige naam'].values)

VASTE_STAF_NAMEN = get_vaste_staf_namelist('../Data/HSR Vaste staf 01-05-2022.XLSX')

def check_if_vaste_staf(naam, vaste_staf_namen=VASTE_STAF_NAMEN):

    return naam in vaste_staf_namen

def check_if_vaste_staf_achternaam(achternaam, vaste_staf_namen=VASTE_STAF_NAMEN):
    vaste_staf_achternamen = [naam.split(' ')[-1].lower() for naam in vaste_staf_namen]

    return achternaam.lower() in vaste_staf_achternamen
```

```python
check_if_vaste_staf_achternaam('Groot')
```

```python
naam_mapping = pd.read_excel('../Data/HSR Vaste staf 01-05-2022.XLSX', sheet_name='NaamMapping')

def supervisornaam_naar_onderzoeksnaam(naam, naam_mapping=naam_mapping):
    try:
        return naam_mapping.loc[naam_mapping.SupervisorNaam == naam, 'Volledige naam'].values[0]
    except IndexError as e:
        print('Error bij naam: |' , naam, '|')
        raise IndexError

```

```python
# sv = supervisie
sv1 = pd.read_excel('../Data/Promovendi HSR 20220602.xlsx', sheet_name='PhD programme')
sv2 = (pd.read_excel('../Data/HSR promoties 01012021 tot 01072022.xlsx', sheet_name='PhD programme')
      .drop(['PhD defense date', 'Duration from PhD starting date until PhD defense date'], axis=1))

assert (sv1.columns.values == sv2.columns.values).all()

sv = pd.concat([sv1, sv2], axis=0)
sv
```

```python
supervisor_df = sv['Assigned supervisors'].str.split(',', expand=True)


ties_prep = pd.concat([sv['Full name'], supervisor_df], axis=1)
ties_prep
```

```python
ties_long = (pd.melt(ties_prep, id_vars='Full name',
                    value_vars=supervisor_df.columns,
                    var_name='supervisor_nr',
                     value_name='Name')
             .sort_values('Full name')
            .rename(columns={'Full name': 'PhD name'})
            .drop('supervisor_nr', axis=1)
            .dropna())

# ties_long = ties_long.dropna()
ties_long
```

```python
ties_long['achternaam'] = ties_long['Name'].str.split().str[-1]
ties_long['is_vaste_staf'] = ties_long['achternaam'].apply(check_if_vaste_staf_achternaam)

# selecteer alleen rijen met vaste stafleden
ties_long = ties_long.loc[ties_long.is_vaste_staf, ]
ties_long
```

```python
vc = ties_long['PhD name'].value_counts().to_frame('nr_of_HSR_supervisors')
vc.index.name='PhD name'

ties = ties_long.merge(vc, on='PhD name', how='left')
ties = ties[ties['nr_of_HSR_supervisors'] > 1]
ties
```

```python
# naam zelfde als tiesnaam maken
ties['Name'] = ties['Name'].str.replace('.', '').str.strip()
ties
```

```python
# pd.DataFrame(ties.sort_values('achternaam').Name.str.strip().unique()).to_csv('../Data/supervisie_namen.csv')
```

```python
# zorg dat namen gelijk geschreven zijn als in Pure/Onderzoek
missing_before = ties['Name'].isnull().sum()
ties['Name'] = [supervisornaam_naar_onderzoeksnaam(naam) for naam in ties['Name'].values]
missing_after = ties['Name'].isnull().sum()
assert missing_before == missing_after
ties
```

```python
# to edges with count

from itertools import combinations
from collections import Counter

all_tuples = []
for phd in ties['PhD name'].unique():
    
    # get rows of that unityear
    # get staff in that unityear
    staff = sorted(ties.loc[ties['PhD name'] == phd, 'Name'].values)

    # get all combinations of 2 of staff in that unityear
    for c in list(combinations(staff, 2)):
        # print(unityear, c)
        all_tuples.append(c)
        
Counter(all_tuples)
        
```
